```python
#Importing Dependencies
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
student_dataset=pd.read_csv('Salary_data.csv',sep=',')
```

```python
predict = 'Salary' 
exp = 'YearsExperience'
X = np.array(student_dataset.drop([predict], axis=1))
y = np.array(student_dataset[predict])
```

```python
x_train, x_test, y_train, y_test=sklearn.model_selection.train_test_split(X,y,test_size=0.2)
Zedex=linear_model.LinearRegression()
Zedex.fit(x_train,y_train)
accuracy=Zedex.score(x_test,y_test)
print(accuracy)
```

```python
print("Coefficients : ",Zedex.coef_)
print("Intercept : ",Zedex.intercept_)
```

```python
predictions=Zedex.predict(x_test)
for i in range(len(predictions)):
    print("Predicted value :",predictions[i],'Actual value :',y_test[i])
```
