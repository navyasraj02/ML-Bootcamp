```python
#importing dependencies
import pandas as pd
import numpy as np
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
```

```python
cars_dataset=pd.read_csv("cars.csv",sep=',')
cars_dataset.head()
```

```python
le = preprocessing.LabelEncoder()     #performing labelEncoding
buying = le.fit_transform(list(cars_dataset['buying']))
maint = le.fit_transform(list(cars_dataset['maint']))
door = le.fit_transform(list(cars_dataset['door']))
persons = le.fit_transform(list(cars_dataset['persons']))
lug_boot= le.fit_transform(list(cars_dataset['lug_boot']))
safety = le.fit_transform(list(cars_dataset['safety']))
cls = le.fit_transform(list(cars_dataset['class']))         #to predict
```

```python
X=list(zip(buying,maint,door,persons,lug_boot,safety))
y=list(cls)
x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(X,y,test_size=0.2)
Friday=KNeighborsClassifier(n_neighbors=9)   #try to take odd nos
Friday.fit(x_train,y_train)
accuracy=Friday.score(x_test,y_test)
print(accuracy*100)
```

```python
predictions=Friday.predict(x_test)


decoder=['unacc','acc','good','vgood']
for i in range(len(predictions)):
    print("Friday predicted : ",decoder[predictions[i]],"Actual value : ",decoder[y_test[i]])
```
